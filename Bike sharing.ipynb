## Treating Dummy Variables:

### Mnth variable

```python
# Applying get_dummies on month variable:
dum_mnth=pd.get_dummies(df['mnth'],drop_first = True)
```

```python
# Renaming month variables:
dum_mnth.rename(columns = {2:'Feb', 3:'March', 4:'April',5:'May', 6:'June', 7:'July',8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}, inplace = True) 

# Adding dummy variables to dataset:
df=pd.concat([df,dum_mnth],axis=1)
```

```python
# Dropping month column:
df.drop(['mnth'], axis = 1, inplace = True)
df.head()
```

###  Weekday variable

```python
# Renaming week days:
dict_wd={0:'sun',1:'mon',2:'tue',3:'wed',4:'thu',5:'fri',6:'sat'}
df['weekday']=df.weekday.map(dict_wd)
```

```python
# applying get dummies on weekdays and dropping one column:
status_wd = pd.get_dummies(df['weekday'], drop_first= True)
status_wd.head()
```

```python
# Adding dummy variables to data frame and removing weekday from the datafarame:
df = pd.concat([df, status_wd], axis = 1)
df.drop(['weekday'], axis=1, inplace=True)
df.head()
```

### For season

```python
# Applying get dummies on season variable:
dum_season=pd.get_dummies(df['season'],drop_first = True)

# Renaming dummy variables in seasons:
dum_season.rename(columns = {2:'summer', 3:'fall', 4:'winter'}, inplace = True) 
```

```python
# Adding variables to the dataframe:
df=pd.concat([df,dum_season],axis=1)

# Dropping season variable:
df.drop(['season'], axis = 1, inplace = True)
df.head()
```

### Weathersit variable

```python
# Applying get dummies on weathersit:
dum_weathersit=pd.get_dummies(df['weathersit'],drop_first = True)

# Renaming dummy variables in weathersit:
dum_weathersit.rename(columns = {2:'Mist', 3:'Light Snow/Rain'}, inplace = True)
```

```python
#  Adding variables to the dataframe:
df=pd.concat([df,dum_weathersit],axis=1)

# Dropping weathersit variable:
df.drop(['weathersit'], axis = 1, inplace = True)
df.head()
```

### Year variable

```python
# Applying get dummies on year:
dum_yr=pd.get_dummies(df['yr'],drop_first = True)

# Renaming dummy variables in year:
dum_yr.rename(columns = {1:'2019'}, inplace = True) 
```

```python
#  Adding variables to the dataframe:
df=pd.concat([df,dum_yr],axis=1)

# Dropping year variable:
df.drop(['yr'], axis = 1, inplace = True)
df.head()
```

### Splitting of Data into Train set and Test set

```python
# Importing train_test_split:
from sklearn.model_selection import train_test_split

# Applying train_test_split on data with test size of 0.3 and random_state of 100:
np.random.seed(0)
df_train, df_test = train_test_split(df, test_size = 0.3, random_state = 100)
```

### Rescaling on the variables

```python
# Importing MinMaxScaler from sklearn.preprocessing:
from sklearn.preprocessing import MinMaxScaler
```

```python
# Creating instance of MinMaxScaller:
scaler = MinMaxScaler()
```

```python
# Creating variable list need scaling:
num_vars = ['atemp','temp', 'hum', 'windspeed','cnt']

# Performing fit.transform on train set:
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_train.head()
```

```python
# Checking distribution of dataframe:
df_train.describe()
```

### Splitting in X_train and y_train sets for building the model

```python
# Creating X_train and y_train sets:
y_train = df_train.pop('cnt')
X_train = df_train
```

### Plotting Heatmap

```python
# Checking correlation between the variables:
plt.figure(figsize = (18, 16))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

### Building Linear Regression model

```python
# Importing RFE and linear regression from sklearn:
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 15
# Creating instance of linear regression as lm:
lm = LinearRegression()
# Fitting model on training set:
lm.fit(X_train, y_train)
```

```python
# Executing RFE wih required variable as 15:
rfe = RFE(lm, 15) 
# Fitting RFE on training set:
rfe = rfe.fit(X_train, y_train)
```

```python
# Getting list of variables with support and ranking:
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# Extracting variables with support = True:
col = X_train.columns[rfe.support_]
col
```

```python
# Dropping variables with support = False:
X_train.columns[~rfe.support_]
```

### Using statsmodel library for detailed statistics

### Iteration 1

```python
# Assigning variables of RFE to X_train:
X_train_rfe = X_train[col]
```

```python
# Importing statsmodel library:
import statsmodels.api as sm  
# Assigning constant:
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
# Running the model:
lm = sm.OLS(y_train,X_train_rfe).fit() 
```

```python
# Printing the summary:
print(lm.summary())
```

#### Checking VIF:

```python
# Importing variance_inflation_factor from statsmodel:
from statsmodels.stats.outliers_influence import variance_inflation_factor

#printing the Vif of model:
vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

 - Workingday is highly correlated with holiday hence let's drop workingday.
 - As the p value of Oct is high hence lets drop Oct variable

### Making Predictions

#### Applying the scaling on the test sets

```python
# Variables on which scaling to be done:
num_vars = ['atemp','temp', 'hum', 'windspeed','cnt']

# Applying scaler transform on selected variables:
df_test[num_vars] = scaler.transform(df_test[num_vars])
```

```python
# Creating X_test and y_test:
y_test = df_test.pop('cnt')
X_test = df_test
```

```python
# Using model to make predictions:

# X_test_new created by dropping columns from X_test: 
X_test_new = X_test[X_train_new.columns]

# Inserting constant in data frame:
X_test_new = sm.add_constant(X_test_new)
```

```python
# Predicting on test set:
y_pred = lm.predict(X_test_new)
```

```python
# Lets plot y_test vs y_pred to see the spread:
fig = plt.figure(figsize=(5,5))
sns.regplot(y_test,y_pred)
plt.ylabel('y_pred', fontsize=18)             
plt.xlabel('y_test', fontsize=16)                          
fig.suptitle('y_test vs y_pred', fontsize=18)
plt.show()
```

```python
# Importing R^2 and MSE from sklearn.metrics:
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

# Calculating RMSE
rsme = np.sqrt(mean_squared_error(y_test,y_pred))
rsme
```

```python
# Calculating R^2:
r_squared = r2_score(y_test,y_pred)
r_squared
```
